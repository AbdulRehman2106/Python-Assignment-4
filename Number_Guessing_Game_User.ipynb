```python
import random
import time
from datetime import datetime
import os


def show_splash_screen():
    os.system('cls' if os.name == 'nt' else 'clear')
    title = """


  _      __    __                      ______         __  __                _  __           __               _____                 _               _____                 __
 | | \| / /__ / /______  __ _  ___    /_  __/__      / / / /__ ___ ____    / |/ /_ ____ _  / /  ___ ____    / ___/_ _____ ___ ___ (_)__  ___ _    / ___/__ ___ _  ___   / /
 | |/ |/ / -_) / __/ _ \/  ' \/ -_)    / / / _ \    / /_/ (_-</ -_) __/   /    / // /  ' \/ _ \/ -_) __/   / (_ / // / -_|_-<(_-</ / _ \/ _ `/   / (_ / _ `/  ' \/ -_) /_/  _ _
 |__/|__/\__/_/\__/\___/_/_/_/\__/    /_/  \___/    \____/___/\__/_/     /_/|_/\_,_/_/_/_/_.__/\__/_/      \___/\_,_/\__/___/___/_/_//_/\_, /    \___/\_,_/_/_/_/\__/ (_)  (_|_)

                                                                 /___/

                                            üî¢ USER  NUMVER  GUESSING  GAME üî¢

                                            Created By Abdul Rehman..
    """
    print(title)
    print("\nüîÑ Loading", end="")
    for _ in range(3):
        time.sleep(0.5)
        print(".", end="", flush=True)
    time.sleep(1)
    print("\n")


def choose_difficulty():
    print("\nüìä Choose Difficulty:")
    print("\n1. Easy (1‚Äì10)")
    print("\n2. Medium (1‚Äì50)")
    print("\n3. Hard (1‚Äì100)")
    print("\n4. Extreme Hard (1‚Äì1000)")

    while True:
        level = input("\nEnter Choice (1/2/3/4): ").strip()
        if level == "1":
            return 1, 10, "Easy"
        elif level == "2":
            return 1, 50, "Medium"
        elif level == "3":
            return 1, 100, "Hard"
        elif level == "4":
            return 1, 1000, "Extreme Hard"
        else:
            print("\n‚ùå Invalid choIce. Please Select 1, 2, 3, or 4.")


def get_secret_number(min_val, max_val):
    while True:
        try:
            secret = int(input(f"\nü§´ Think Of A Number Between {min_val} And {max_val}: "))
            if min_val <= secret <= max_val:
                return secret
            else:
                print("\n‚ö†Ô∏è Out Of Range.")
        except:
            print("\n‚ùå Invalid Input. Enter A Number.")


def computer_guess(secret, min_val, max_val):
    low = min_val
    high = max_val
    guesses = []
    attempts = 0

    while low <= high:
        guess = (low + high) // 2
        guesses.append(guess)
        attempts += 1
        time.sleep(1)
        print(f"\nüí° Computer's Guess #{attempts}: {guess}")

        if guess == secret:
            print("\n‚úÖ‚úÖüéâ Computer Guessed Correctly!")
            break
        elif guess > secret:
            print("\nüîΩ Too high.")
            high = guess - 1
        else:
            print("\nüîº Too low.")
            low = guess + 1

    print(f"\nüìú Guess History: {guesses}")
    return attempts


def players_guess(min_val, max_val):
    number = random.randint(min_val, max_val)
    attempts = 0

    while True:
        try:
            guess = int(input(f"\nüéØ Your Guess ({min_val}-{max_val}): "))
            attempts += 1

            if guess < number:
                print("\nüîº Too Low!")
            elif guess > number:
                print("\nüîΩ Too High!")
            else:
                print("\n‚úÖ Correct!")
                break
        except ValueError:
            print("\n‚ùå Invalid Input. Enter A Number.")
    return attempts


def save_score(name, mode, level, attempts):
    with open("score_history.txt", "a") as f:
        f.write(f"\n{datetime.now()}, {name}, {mode}, {level}, {attempts} attempts\n")


def show_leaderboard():
    try:
        with open("score_history.txt", "r") as f:
            lines = f.readlines()[-5:]
        print("\nüèÜ Leaderboard (Recent Games):")
        for line in lines:
            print("‚Ä¢", line.strip())
    except FileNotFoundError:
        print("\nüìÇ No Score History Yet.")


def main():
    print("\nüéÆ WELCOME TO THE USER NUMBER GUESSING GAME üéÆ".center(130))
    show_leaderboard()
    name = input("\nEnter Your Name: ")

    while True:
        print("\nüéÆ Game Modes:")
        print("\n1. Computer Guesses Your Number")
        print("\n2. You Guess Computer's Number")
        print("\n3. Exit")
        choice = input("\nChoose Mode (1/2/3): ")

        if choice == "1":
            min_val, max_val, level = choose_difficulty()
            secret = get_secret_number(min_val, max_val)
            attempts = computer_guess(secret, min_val, max_val)
            save_score(name, "Computer Guessed", level, attempts)

        elif choice == "2":
            min_val, max_val, level = choose_difficulty()
            attempts = players_guess(min_val, max_val)
            save_score(name, "Player Guessed", level, attempts)

        elif choice == "3":
            print("\nüëã Thanks For Playing! Goodbye.")
            break
        else:
            print("\n‚ùå Invalid Choice.")


show_splash_screen()
main()
```
