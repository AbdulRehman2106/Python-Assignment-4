```python
import random
import time
import sys
import os

choices = ["Rock", "Paper", "Scissors"]



def show_splash_screen():
    os.system('cls' if os.name == 'nt' else 'clear')

    title = """


  _      __    __                        ______           ___           __        ___                         ____    _                           _____                 __
 | | /| / /__ / /______  __ _  ___      /_  __/__        / _ \___  ____/ /__     / _ \___ ____  ___ ____     / __/___(_)__ ___ ___  _______      / ___/__ ___ _  ___   / /
 | |/ |/ / -_) / __/ _ \/  ' \/ -_)      / / / _ \      / , _/ _ \/ __/  '_/    / ___/ _ `/ _ \/ -_) __/    _\ \/ __/ (_-<(_-</ _ \/ __(_-<     / (_ / _ `/  ' \/ -_) /_/ _
 |__/|__/\__/_/\__/\___/_/_/_/\__/      /_/  \___/     /_/|_|\___/\__/_/\_\    /_/   \_,_/ .__/\__/_/      /___/\__/_/___/___/\___/_/ /___/     \___/\_,_/_/_/_/\__/ (_|_|_)
                                                                                        /_/



                                                    ðŸŽ® ROCK PAPER SCISSORS GAME ðŸŽ®


                                                    Created By Abdul Rehman..
    """
    print(title)
    print("\n" + "ðŸ”„ Loading", end="")
    for _ in range(3):
        time.sleep(1)
        print(".", end="", flush=True)
    time.sleep(1)
    print("\n")





def print_score(user_score, computer_score):
    print(f"\nYour Score: {user_score} | Computer Score: {computer_score}")

def typing_animation(text, delay=0.1):
    for char in text:
        sys.stdout.write(char)
        sys.stdout.flush()
        time.sleep(delay)
    print()

def bonus_round():
    typing_animation("\nðŸŽ¯ It's A Tie! Bonus Round To Determine The Winner ðŸŽ¯", 0.05)
    user_choice = input("\nRock, Paper, or Scissors?: ").capitalize()

    if user_choice not in choices:
        typing_animation("\nâŒâŒâŒ Invalid choice. Please choose Rock, Paper, or Scissors.", 0.05)
        return bonus_round()

    computer_choice = random.choice(choices)
    typing_animation(f"\nComputer Choose: {computer_choice}", 0.05)

    if user_choice == computer_choice:
        typing_animation(f"Match Draw! ðŸ¤ Both Choose The Same!", 0.05)
        return 0
    elif (user_choice == "Rock" and computer_choice == "Scissors") or \
         (user_choice == "Paper" and computer_choice == "Rock") or \
         (user_choice == "Scissors" and computer_choice == "Paper"):
        typing_animation("You Win the Bonus Round! ðŸŽ‰", 0.05)
        return 1
    else:
        typing_animation("\nComputer Wins the Bonus Round!", 0.05)
        return -1

def play_game():
    user_score = 0
    computer_score = 0
    rounds = 3

    typing_animation("\nðŸŽ® Rock, Paper, Scissors Game ðŸŽ®", 0.05)

    for i in range(rounds):
        typing_animation(f"\nRound {i + 1}: \n", 0.05)
        user_choice = input("Rock, Paper, or Scissors?: ").capitalize()

        if user_choice not in choices:
            typing_animation("\nâŒâŒâŒ Invalid Choice. Please Choose Rock, Paper, Or Scissors.", 0.05)
            continue

        computer_choice = random.choice(choices)
        typing_animation(f"\nComputer Choose: {computer_choice}", 0.05)

        if user_choice == computer_choice:
            typing_animation(f"Match Draw! ðŸ¤ Both Choose The Same!", 0.05)
        elif (user_choice == "Rock" and computer_choice == "Scissors") or \
             (user_choice == "Paper" and computer_choice == "Rock") or \
             (user_choice == "Scissors" and computer_choice == "Paper"):
            typing_animation("\nYou Win This Round! ðŸŽ‰", 0.05)
            user_score += 1
        else:
            typing_animation("\nComputer Wins This Round!", 0.05)
            computer_score += 1

        print_score(user_score, computer_score)
        time.sleep(1)

    if user_score == computer_score:
        typing_animation("\nðŸŽ¯ It's a Tie after 3 rounds! Time for the Bonus Round!", 0.05)
        bonus_result = bonus_round()

        if bonus_result == 1:
            user_score += 1
        elif bonus_result == -1:
            computer_score += 1

    if user_score > computer_score:
        typing_animation("\nðŸŽ‰ Congratulations! You are the overall winner!", 0.05)
    elif computer_score > user_score:
        typing_animation("\nComputer Wins the Game!", 0.05)
    else:
        typing_animation("\nIt's a tie game!", 0.05)

    print_score(user_score, computer_score)

    play_again = input("\nDo You Want To Play Again? (Yes/No): ").capitalize()
    if play_again != "Yes":
        typing_animation("Thanks For Playing! ðŸ˜Š", 0.05)
        return

    play_game()

show_splash_screen()
play_game()
```
